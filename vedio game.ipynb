```python
#Analysing yearly sales of 7th Gen Consoles
video7th = df[(df['Platform'] == 'Wii') | (df['Platform'] == 'PS3') | (df['Platform'] == 'X360')]
yearlySales = video7th.groupby(['Year_of_Release','Platform']).Global_Sales.sum()
yearlySales.unstack().plot(kind='bar',grid=False,figsize=(10,7))
plt.title('Stacked Barplot of Global Yearly Sales of the 7th Gen Consoles')
plt.ylabel('Global Sales')
```

```python
# we have a lot of games 
print(df['Publisher'].value_counts(dropna =False))
```

```python
# vedio games on different platforms
plt.figure(figsize=(20,10))
sns.countplot(df['Platform'])
plt.title('Number of games on each platform')
plt.xlabel('Platform')
plt.ylabel('Number of games')
```

```python
#groupby platform max sales of which platform
sd=df.groupby('Platform').agg({'Global_Sales':'sum'})
plt.figure(figsize=(20,10))
sns.barplot(y='Global_Sales',x=sd.index,data=sd)
```
